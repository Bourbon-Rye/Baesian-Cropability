```python
import pandas as pd
import os
import glob 
#import numpy as np

files_list = os.path.join("C:/Users/Ralph/Downloads/UP 2nd Sem 23-24/CS 132/dataset/Bourbon-Rye Baesian-Cropability data1 datasets-prices_cpi", "cpi_corecpiforallincomehouseholdsbycg*.csv") 
files_list = glob.glob(files_list) 
print(files_list)

files_list.sort()
f = files_list.pop(0)
df = pd.read_csv(f)

for nfile in files_list:
    print(nfile)
    nfile = pd.read_csv(nfile)
    cols_to_use = list(nfile.columns.intersection(df.columns))
    samp = list(nfile.columns.difference(df.columns))
    print(cols_to_use)
    df.replace('..', 0.0, inplace=True)
    nfile.replace('..', 0.0, inplace=True)

    df = pd.merge(df, nfile[samp],  left_index=True, right_index=True, how='outer')
    for i in range (len(cols_to_use)-1):
        df[cols_to_use[i]] = df[cols_to_use[i]].fillna(nfile[cols_to_use[i]])

print(df)
df.to_csv('cpi_corecpiforallincomehouseholdsbycg.csv')
```
